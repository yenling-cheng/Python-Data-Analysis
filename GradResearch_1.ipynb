```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_NA=pd.read_csv('NorthAmerica.csv')
df_BNT=pd.read_csv('BNT.csv')
df_DAP=pd.read_csv('Developed Asia Pacific.csv')
df_DE=pd.read_csv('Developed Europe.csv')
df_EAP=pd.read_csv('Emerging Asia Pacific.csv')
df_EE=pd.read_csv('Emerging Europe.csv')
df_LA=pd.read_csv('Latin America.csv')
df_ME=pd.read_csv('Middle East.csv')
df_Afr=pd.read_csv('Africa.csv')


df_NA['datadate']=pd.to_datetime(df_NA['datadate'], format="%Y%m%d")
df_BNT['datadate']=pd.to_datetime(df_BNT['datadate'],format="%Y/%m/%d")
df_DAP['datadate']=pd.to_datetime(df_DAP['datadate'],format="%Y/%m/%d")
df_DE['datadate']=pd.to_datetime(df_DE['datadate'],format="%Y/%m/%d")
df_EAP['datadate']=pd.to_datetime(df_EAP['datadate'],format="%Y/%m/%d")
df_EE['datadate']=pd.to_datetime(df_EE['datadate'],format="%Y/%m/%d")
df_LA['datadate']=pd.to_datetime(df_LA['datadate'],format="%Y/%m/%d")
df_ME['datadate']=pd.to_datetime(df_ME['datadate'],format="%Y/%m/%d")
df_Afr['datadate']=pd.to_datetime(df_Afr['datadate'],format="%Y/%m/%d")

df_NA
```

```python
df_NA['Daily Return']=0.0
for i in range (1,len(df_NA)):
    df_NA.at[i,'Daily Return']=float(np.log2(df_NA.at[i,'prccd']))-float(np.log2(df_NA.at[i-1,'prccd']))

df_BNT['Daily Return']=0.0
for i in range (1,len(df_BNT)):
    df_BNT.at[i,'Daily Return']=float(np.log2(df_BNT.at[i,'prccd']))-float(np.log2(df_BNT.at[i-1,'prccd']))

df_DAP['Daily Return']=0.0
for i in range (1,len(df_DAP)):
    df_DAP.at[i,'Daily Return']=float(np.log2(df_DAP.at[i,'prccd']))-float(np.log2(df_DAP.at[i-1,'prccd']))
    
df_DE['Daily Return']=0.0
for i in range (1,len(df_DE)):
    df_DE.at[i,'Daily Return']=float(np.log2(df_DE.at[i,'prccd']))-float(np.log2(df_DE.at[i-1,'prccd']))
    
df_EAP['Daily Return']=0.0
for i in range (1,len(df_EAP)):
    df_EAP.at[i,'Daily Return']=float(np.log2(df_EAP.at[i,'prccd']))-float(np.log2(df_EAP.at[i-1,'prccd']))
    
df_EE['Daily Return']=0.0
for i in range (1,len(df_EE)):
    df_EE.at[i,'Daily Return']=float(np.log2(df_EE.at[i,'prccd']))-float(np.log2(df_EE.at[i-1,'prccd']))
    
df_LA['Daily Return']=0.0
for i in range (1,len(df_EE)):
    df_LA.at[i,'Daily Return']=float(np.log2(df_LA.at[i,'prccd']))-float(np.log2(df_LA.at[i-1,'prccd']))
    
    
df_ME['Daily Return']=0.0
for i in range (1,len(df_ME)):
    df_ME.at[i,'Daily Return']=float(np.log2(df_ME.at[i,'prccd']))-float(np.log2(df_ME.at[i-1,'prccd']))
 
    
df_Afr['Daily Return']=0.0
for i in range (1,len(df_Afr)):
    df_Afr.at[i,'Daily Return']=float(np.log2(df_Afr.at[i,'prccd']))-float(np.log2(df_Afr.at[i-1,'prccd']))

df_NA
df_BNT
```

```python
ax=plt.gca()
df_NA.plot(kind='line', title='North America', x='datadate', y='Daily Return',ax=ax, linewidth=0.25)
plt.show()

ax=plt.gca()
df_BNT.plot(kind='line', title='BioNTech', x='datadate', y='Daily Return',ax=ax, linewidth=0.5)
plt.show()

ax=plt.gca()
df_DAP.plot(kind='line', title='Developed Asia Pacific', x='datadate', y='Daily Return',ax=ax, linewidth=0.25)
plt.show()

ax=plt.gca()
df_DE.plot(kind='line', title='Developed Europe', x='datadate', y='Daily Return',ax=ax, linewidth=0.25)
plt.show()

ax=plt.gca()
df_EAP.plot(kind='line', title='Emerging Asia Pacific', x='datadate', y='Daily Return',ax=ax, linewidth=0.25)
plt.show()

ax=plt.gca()
df_EE.plot(kind='line', title='Emerging Europe', x='datadate', y='Daily Return',ax=ax, linewidth=0.25)
plt.show()

ax=plt.gca()
df_LA.plot(kind='line', title='Latin America', x='datadate', y='Daily Return',ax=ax, linewidth=0.25)
plt.show()

ax=plt.gca()
df_ME.plot(kind='line', title='Middle East', x='datadate', y='Daily Return',ax=ax, linewidth=0.5)
plt.show()

ax=plt.gca()
df_Afr.plot(kind='line', title='Africa', x='datadate', y='Daily Return',ax=ax, linewidth=0.25)
plt.show()
```

```python
from statsmodels.tsa.stattools import grangercausalitytests
BNT_NA=pd.merge((df_NA.loc[:,['datadate','Daily Return']]).rename(columns={'Daily Return':'North America'}),\
           (df_BNT.loc[:,['datadate','Daily Return']]).rename(columns={'Daily Return':'BNT'}), on='datadate', how='inner')
BNTvsNA=BNT_NA.drop(columns=['datadate'])
BNT_NA_result=grangercausalitytests(BNTvsNA, 5)
```
